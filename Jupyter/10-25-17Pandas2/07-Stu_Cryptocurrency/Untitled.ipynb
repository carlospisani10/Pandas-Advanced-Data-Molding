```python
import pandas as pd
```

```python
bitpath = "Resources/bitcoin_cash_price.csv"
bitcoin = pd.read_csv(bitpath)

dashpath = "Resources/dash_price.csv"
dash = pd.read_csv(dashpath)
```

```python
merge_table = pd.merge(bitcoin, dash, on ="Date",how="outer")
```

```python
merge_table.head()
```

```python
col_change = merge_table.rename(columns={"Open_x" : "Bitcoin Open:", "Volume_x" : "Bitcoin Volume", "Market Cap_x" : "Market Cap Bitcoin", "Open_y" : "Dash Open", "Volume_y" : "Dash Volume", "Market Cap_y" : "Market Cap Dash"})
col_change.head()
```

```python
# COllecting best open for bitcoin and dash
bitcoin_open = col_change["Bitcoin Open:"].max()
bitcoin_open.head()
```
